Imports the packages that we need

```python
from fastai.vision.all import *
from fastai.vision.widgets import *
from jmd_imagescraper.core import *
```

Creates `katy_zoe` folder in the local directory. Downloads images for each actress under `katy_zoe/katy perry` and `katy_zoe/zooey deschanel`

```python
katy_zoe = 'katy perry', 'zooey deschanel'
path = Path('katy_zoe')

if not path.exists():
    path.mkdir()
    for actress in katy_zoe:
        duckduckgo_search(path, actress, actress)
```

## Check if images are downloaded correctly

Sometimes images we download are corrupted. We get the list of all the images we downloaded and we delete those that are corrupted

```python
fns = get_image_files(path)
fns
```

```python
len(fns)
```

```python
failed = verify_images(fns)
failed
```

```python
failed.map(Path.unlink)
```

## DataBlock and DataLoaders and Image Augmentation

`DataLoaders` are objects that you pass data to and allow you to loop over the segregated `train` and `valid` data.

`DataBlock` defines the specs of the `DataLoader`

```python
actresses = DataBlock(
    blocks=(ImageBlock, CategoryBlock), #Specify the independent variable (Images) and dependent variable (Categories/Classes)
    get_items=get_image_files, # Specify how to get the data
    splitter=RandomSplitter(valid_pct=0.2, seed=42), # Specify how to split training and validation data
    get_y=parent_label, # The label is specified as the folder that contains the image
    item_tfms=Resize(128), # Resize each image to 128x128 pixels
    batch_tfms=aug_transforms(mult=2) # Specify transforms that are commonly used for image classification tasks
)
```

```python
dls = actresses.dataloaders(path)
dls.train.show_batch(max_n=4, nrows=1)
```

## Training the model by finetuning a `resnet18` model

```python
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

Show which images our model performs poorly

```python
interp.plot_top_losses(5, nrows=1)
```

```python
cleaner = ImageClassifierCleaner(learn)
cleaner
```

```python
for idx in cleaner.delete(): cleaner.fns[idx].unlink()
for idx, cat in cleaner.change(): shutil.move(str(cleaner.fns[idx]), path/cat)
```

```python
dls = actresses.dataloaders(path)
learn = cnn_learner(dls, resnet18, metrics=error_rate)
```

```python
learn.fine_tune(10)
```

```python
interp = ClassificationInterpretation.from_learner(learn)
interp.plot_confusion_matrix()
```

```python
interp.plot_top_losses(3, nrows=1)
```

This will export the model to a `export.pkl` file in the current directory

```python
learn.export()
```

Check exported model exists

```python
path = Path()
path.ls(file_exts='.pkl')
```

Reload the model from the exported file

```python
learn_inf = load_learner(path/'export.pkl')
```

Set-up IPython Widgets for us to test our model

```python
btn_upload = widgets.FileUpload()
lbl_pred = widgets.Label()
btn_run = widgets.Button(description='Classify')
out_pl = widgets.Output()

def on_click_classify(change):
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; Probability: {probs[pred_idx]:.04f}'

btn_run.on_click(on_click_classify)
```

```python
VBox([widgets.Label('Upload a picture!'), 
      btn_upload, 
      btn_run, 
      out_pl, 
      lbl_pred])
```
